# Improving Animal Protection in the Mountain West with Convolutional Neural Networks 

The goal of this project is to decrease dangerous animal/human interactions in the Mountain West using image recognition of animal species. Every year, [hundreds of animals](https://www.vaildaily.com/news/eagle-valley/colorado-sees-slight-drop-in-bear-reports-in-unfortunately-normal-year/) in states like Colorado are euthanized following human interactions. Even if a human is not injured or killed, frequent human interactions desensitize animals to humans' presence, usually resulting in a decision to euthanize the animal for human safety. Often when humans come into contact with the animals, they get too close, take pictures, and leave food and trash out. While it's exciting for humans to see a bear or moose, these actions are often fatal for the animal. 

Using camera traps around campsites, tourist attractions, and other highly-trafficked sites, park rangers can identify the presence of animals that could possibly injure humans - bears, moose, cougars, and coyotes - and step in before an interaction occurs. 

For this project, I'm using a small subset of [North American Camera Trap Images from the Labeled Information Library of Alexandria](https://lila.science/datasets/nacti). It will be a binary task to flag 'dangerous' animals while bypassing others. I selected four species to be marked 'dangerous' - american black bear, moose, cougar, coyote - and six categories that are common but not cause usually for concern : squirrel, deer, elk, raccoon, chipmunk, and empty. So far, I've run three models on this data:
- Random Forest model: Accuracy 0.81, Precision 0.84, Recall 0.85
- Basic CNN model with 10 epochs: Accuracy 0.64, Precision 0.64, Recall 0.89
- CNN model built using MobileNetV2: Accuracy 0.86, Precision 0.9, Recall 0.86

Moving forward, I plan to tune the transfer learning model with tools like early stopping, drop layers, and tinkering with hyperparameters. Finally, I'd like to attempt object detection using bounding boxes.
